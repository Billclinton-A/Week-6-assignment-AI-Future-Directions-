```python
# Install graphviz if needed
!apt-get install -y graphviz
!pip install graphviz

from graphviz import Digraph

dot = Digraph(comment='AI-Driven IoT Smart Agriculture')

# Nodes
dot.node('S', 'IoT Sensors\n(Soil, Temp, Humidity, Light, pH, EC, Rainfall, Camera)')
dot.node('E', 'Edge Device / Microcontroller\n(Raspberry Pi)')
dot.node('C', 'Cloud Server / AI Model\n(LSTM / XGBoost)')
dot.node('P', 'Prediction & Analytics\n(Crop Yield, Irrigation Advice)')
dot.node('A', 'Optional Actuators\n(Irrigation, Fertilizer)')
dot.node('D', 'Visualization / Dashboard\n(Plots, Alerts)')
dot.node('DS', 'Optional Data Storage / IoT Gateway')

# Edges
dot.edges(['SE', 'EC'])
dot.edge('E', 'DS', style='dashed', label='Optional Storage')
dot.edge('C', 'P')
dot.edge('P', 'A', style='dashed', label='Optional Automation')
dot.edge('P', 'D')

# Render diagram
dot.format = 'png'
dot.render('ai_iot_smart_agriculture', view=True)
```
